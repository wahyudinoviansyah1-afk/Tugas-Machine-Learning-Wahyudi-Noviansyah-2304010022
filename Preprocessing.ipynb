# About Dataset

Dataset Air Quality Data in India yang tersedia di Kaggle berisi informasi mengenai kondisi kualitas udara di berbagai kota di India dalam rentang waktu sekitar tahun 2015 hingga 2020. Data ini dikumpulkan oleh Central Pollution Control Board sebagai lembaga resmi pemerintah yang bertugas memantau dan mengendalikan tingkat pencemaran udara. Secara umum, dataset ini merepresentasikan bagaimana kualitas udara berubah dari waktu ke waktu akibat aktivitas manusia seperti transportasi, industri, dan faktor lingkungan lainnya, sehingga sangat relevan untuk analisis lingkungan maupun kesehatan masyarakat.

Data disajikan dalam bentuk harian untuk setiap kota, yang mencakup informasi tanggal dan lokasi serta berbagai parameter polutan udara. Variabel-variabel tersebut meliputi konsentrasi partikel seperti PM2.5 dan PM10, serta gas berbahaya seperti karbon monoksida (CO), nitrogen dioksida (NO2), sulfur dioksida (SO2), ozon (O3), dan beberapa senyawa organik volatil seperti benzena dan toluena. Seluruh parameter ini biasanya diukur dalam satuan mikrogram per meter kubik (µg/m³) dan digunakan untuk menggambarkan tingkat pencemaran udara pada hari tertentu di suatu kota.

Selain itu, dataset ini juga menyediakan nilai Air Quality Index (AQI) yang merupakan indikator utama untuk menilai kualitas udara secara keseluruhan. AQI dihitung berdasarkan beberapa polutan utama, di mana nilai tertinggi dari sub-indeks polutan digunakan sebagai representasi kondisi udara. Dataset juga menyertakan kategori kualitas udara dalam bentuk AQI_Bucket, seperti Good, Moderate, Poor, hingga Severe.

# Cleaning Data

```python
df['Date'] = pd.to_datetime(df['Date'])

df = df.sort_values('Date')
df = df.set_index('Date')
```

```python
df = df.drop_duplicates()
```

Baris yang tidak memiliki informasi signifikan (seluruh kolom numerik kosong) dihapus untuk menjaga kualitas data yang digunakan.

```python
cols_to_check = df.columns.difference(['City'])
df = df.dropna(subset=cols_to_check, how='all')
```

# Handling Nilai 0

Pada beberapa variabel tertentu, nilai nol dianggap tidak valid sehingga dikonversi menjadi missing value (NaN) agar dapat ditangani dengan metode imputasi yang sesuai

```python
cols_invalid_zero = ['Benzene', 'Toluene', 'CO', 'Xylene', 'NOx']
df[cols_invalid_zero] = df[cols_invalid_zero].replace(0, np.nan)
```

# Handling Missing Value

Kolom dalam dataset dipisahkan menjadi dua kategori, yaitu kolom numerik dan kolom kategorikal, untuk mempermudah proses preprocessing yang sesuai dengan karakteristik masing-masing data.

```python
num_cols = df.select_dtypes(include=['float64','int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

Nilai yang hilang pada kolom numerik diisi menggunakan metode interpolasi linear dan median, sedangkan pada kolom kategorikal diisi menggunakan nilai modus. Pendekatan ini bertujuan untuk mempertahankan distribusi data.

```python
df[num_cols] = df[num_cols].interpolate(method='linear')
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

# Handling Outliers (IQR)

Outlier ditangani menggunakan metode Interquartile Range (IQR) dengan teknik capping, yaitu membatasi nilai ekstrem agar berada dalam rentang yang wajar tanpa menghapus data.

```python
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[column] = df[column].clip(lower, upper)
    return df

for col in num_cols:
    df = handle_outliers_iqr(df, col)
```

# Feature Engineering

Dilakukan ekstraksi fitur baru dari data tanggal, yaitu tahun dan bulan untuk menambah informasi yang dapat digunakan dalam analisis atau pemodelan.

```python
df['year'] = df.index.year
df['month'] = df.index.month
```

# Split Data

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)

print(f"Shape of training data: {df_train.shape}")
print(f"Shape of testing data: {df_test.shape}")
```

```python
output_file = 'train_test_data.xlsx'
with pd.ExcelWriter(output_file) as writer:
    df_train.to_excel(writer, sheet_name='Train_Data', index=True)
    df_test.to_excel(writer, sheet_name='Test_Data', index=True)
print(f"Data train dan test telah disimpan ke '{output_file}' dengan sheet terpisah.")
```
